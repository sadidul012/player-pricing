```python
def split_price_history(x):
    x = x[1:-1].split(",")
    x = x if len(x) > 1 else []
    rows = []
    for i in range(0, len(x), 3):
        rows.append([(",".join(x[i:i+2])).strip()[1:-1], float(x[i+2])])
    return rows
```

```python
pc_prices = dataset_2022[["PC dialy prices"]]
pc_prices["PC dialy prices"] = pc_prices["PC dialy prices"].apply(lambda x: split_price_history(x))
pc_prices.head()
```

```python
pc_prices["pc_price_len"] = pc_prices["PC dialy prices"].apply(lambda x: len(x))
print(pc_prices["pc_price_len"].max())
pc_prices.head()
```

```python
def float_convert(x, index):
    try:
        x = x[1:-1].split("',")[index][1:].replace(",", "")
        return float(x)
    except Exception as e:
        return 0
```

```python
pc_games = dataset_2022['GAMES(PS,Xbox,PC)'].apply(lambda x: float_convert(x, 0))
pc_games.head(10)
```

```python
pc_games[pc_games > 0].shape
```

```python
pc_goals = dataset_2022['GOALS(PS,Xbox,PC)'].apply(lambda x: float_convert(x, 0))
pc_goals.head(10)
```

```python
pc_assists = dataset_2022['ASSISTS(PS,Xbox,PC)'].apply(lambda x: float_convert(x, 0))
pc_assists.head(10)
```
